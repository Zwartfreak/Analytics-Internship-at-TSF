# To Explore Decision Tree Algorithm

```python
# Importing necessary algorithms

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

## Step 1: Loading Data

```python
data = pd.read_csv('iris.csv')
```

```python
data.head()
```

```python
data = data.drop('Id', axis=1)
```

## Step 2: Data Prprocessing

```python
data.isnull().sum().sum()
```

```python
data.dtypes
```

## Step 3: Data Modelling

```python
y = data['Species']
X = data.drop('Species', axis=1)
```

```python
dt = DecisionTreeClassifier()
model = dt.fit(X,y)
```

## Step 4: Visualizing Decision Trees

We can visualize `Decision Trees` using only a single Python library `tree` of `sklearn`.

```python
text = tree.export_text(dt)
```

```python
print(text)
```

```python
with open('decision_tree.log', 'w')as log:
    log.write(text)
```

```python
fig = plt.figure(figsize=(20,10))
tree.plot_tree(dt, feature_names=data.columns, class_names=data['Species'] ,filled=True)
```
